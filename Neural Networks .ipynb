```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python

df = pd.read_csv('iris.data')
df

```

```python
df.columns = ['SepalLengthcm', 'SepalWidthcm', 'PetalLengthcm', 'PetalWidthcm', 'Species']
df
```

```python
df.info()
```

```python
x_input =  df.drop(columns ='Species')
y_input = df['Species']
from sklearn.preprocessing import LabelEncoder
y_encoded  = LabelEncoder()
y_encoded.fit_transform(y_input)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x_input,y_input, test_size = 0.3)
```

```python
# Build a shallow network model
from keras.models import Sequential
from keras.layers import Dense
```

```python
model = Sequential()
model.add(Dense(4, input_shape =(4,),activation = 'relu'))
model.add(Dense(3, activation = 'softmax'))
model.add(Dense(5, activation = 'relu'))
```

```python
model.summary()
```

```python
initial_weight = model.get_weights()
initial_weight
```

```python
model.compile(optimizer = 'adam', loss = 'sparse_categorical_crossentropy', metrics = ['accuracy'])
```

```python
model.fit(x_train, y_train, epochs = 100)
```
