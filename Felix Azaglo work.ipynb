# Filling all null cells with mean values of the already filled cells
### This action is taken so that data is not lost since the data is always better kept than deleted for reasons unforeseen in the future.Hence the null cells are filled with mean of all values in oder to establish correlation with existing data in the cells
## Filling null cells with mean values is a common technique used in data analysis to handle missing data. This is because null or missing data can cause problems in statistical analysis and can result in biased or incomplete results.

## By filling the null cells with the mean of all values, we are preserving the statistical properties of the data, such as the mean and standard deviation. This helps to maintain the overall distribution of the data and ensures that the analysis is based on the most complete dataset possible.

## Additionally, filling null cells with mean values can also help to reduce the impact of outliers or extreme values, which can skew the results of the analysis. By replacing null cells with the mean, we are essentially smoothing out the dataset and making it more representative of the underlying distribution.

```python
everyColumn = ['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar', 'chlorides','free sulfur dioxide', 'total sulfur dioxide','density', 'pH', 'sulphates', 'alcohol', 'quality']
```

```python
for column in everyColumn:
    data[column].fillna(data[column].mean(), inplace=True)
```

#   Correlation between other quantitites and quality

```python
acids = ['fixed acidity', 'citric acid', 'volatile acidity']
```

```python
for acid in acids :
    plt.title(f'Correlation between {acid} and the pH')
    sbn.scatterplot(x=acid, y = 'pH', data=data)
    plt.figure(figsize=(12,10))
    plt.show()
```

```python

columns = ['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar', 'chlorides',
           'free sulfur dioxide', 'total sulfur dioxide', 'density', 'pH', 'sulphates', 'alcohol', 'quality']

for column in columns:
    data[column].mean()
    comparing_means = data[column] < data[column].mean()
    colors = ['#1f77b4', '#2ca02c'] # blue and green colors
    plt.pie(comparing_means.value_counts(), labels=['Values less than the mean', 'Values greater than the mean'],
            colors=colors)
    plt.title(f'Comparison between {column} values and the {column} mean')
    plt.show()


```

# Data correlation using Heat Maps

```python
correlation_matrix = data.corr()

my_cmap = sbn.color_palette(['#bcbddc', '#9e9ac8', '#807dba', '#6a51a3', '#4a1486'])


sbn.heatmap(correlation_matrix, annot=True, cmap=my_cmap)

plt.figure(figsize=(13, 6))

plt.show()
```

```python
sbn.kdeplot(data=data, shade=True)
plt.show
fig, axes = plt.subplots(nrows=len(data.columns), ncols=1, figsize=(10, 16))
for i, col in enumerate(data.columns):
    sbn.kdeplot(ax=axes[i], data=data[col], shade=True)
    axes[i].set_xlabel(col)

plt.tight_layout()
plt.show()
```
